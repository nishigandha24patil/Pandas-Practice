# DataFrames
When you Combine Multiple Series

```python
import numpy as np
import pandas as pd
```

# Creating a DataFrame

```python
data = {
    'Name' : ['Elena', 'Mike', 'Bean', 'John'],
    'Age' : [22, 34, 43, 36],
    'City' : ['London', 'Paris', 'Berlin', 'New York'],
    'Salary' : [90000, 80000, 75000, 88000]
} # dictionary
```

```python
df = pd.DataFrame(data)
df
```

```python
list1 = [
    ['Elena', 22, 'London', 90000],
    ['Mike', 34, 'Paris', 80000],
    ['Bean', 43, 'Berlin', 75000],
    ['John', 36, 'New York', 88000]
] # list
df2 = pd.DataFrame(list1)
df2
```

```python
columns = ['Name', 'Age', 'City', 'Salary']
df3 = pd.DataFrame(list1, columns = columns)
df3
```

# Selection and Indexing of Columns

```python
df3['Name']
```

```python
df3['Salary']
```

### Creating a New Column

```python
df3
```

```python
df3["Designation"] = ['Doctor', 'Engineer', 'Pediatrician', 'Accountant']
```

```python
df3
```

### Removing Columns

```python
df3.drop('Designation', axis = 1, inplace = True) 
# by default axis = 0 (Horizontal)
# inplace = True means Removes the content from the main data
```

```python
df3
```

```python
df3.drop(['City', 'Salary'], axis = 1) # Deletes  Vertical Columns
```

```python
df3.drop(0, axis = 0) # Horizontal rows
```
