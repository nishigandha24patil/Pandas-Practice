# Pandas Series
A series is a one-dimensional labeled array capable of holding any data type. The axis labels are collectively called the index
Series : Vertical
Array : Horizontal

# Creating Series
There are multiple ways to create a Pandas Series : 

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a','b','c']
l1 = [10,20,30]
arr = np.array([10,20,30])
d1 = {'A 100, 2:200, 3:300}
```

```python
pd.Series(l1)
```

```python
pd.Series(l1,index = labels)
```

```python
pd.Series(arr, index = l1) # can't give 2D array because series id 1D
```

```python
pd.Series(d1)
```
